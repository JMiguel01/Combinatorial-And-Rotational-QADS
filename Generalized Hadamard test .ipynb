```python
import math
import numpy as np

import matplotlib.pyplot as plt
xaxisfigure21 = np.arange(0,1.8,0.2)
xaxisfigure22 = np.arange(1.6,3.4,0.2)
```

```python
def probabilityq(angle,m):
    z = (math.cos(angle/2))**(2*m)
    return(z)
def bernoulli(n, q):
    n_success = 0
    for i in range(n):
        random_number = np.random.random()
        if random_number < q:
            n_success += 1
    return(n_success)
def anglep(a,m):
    p = a/(10**4)
    alpha = math.acos(2*(p**(1/m))-1) 
    #alpha = 2*math.acos(p**(1/(2*m))) 
    return(alpha)
def mae(beta,n,m):
    alphas = []
    for i in range(n):
        alphas.append(abs(anglep(bernoulli(10**4, (probabilityq(beta,m))),m)-beta))
    #print(alphas)
    Sum = sum(alphas)
    #print(Sum)
    result = Sum/n
    #result = abs((Sum)/n)
    return(result)
def yaxis(xaxis,m):
    yaxisfigure = []
    for i in xaxis:
        yaxisfigure.append(mae(i,10**3,m))
    return(yaxisfigure)
```

```python
plt.plot(xaxisfigure21, yaxis(xaxisfigure21,1), color='blue', label = "$m=1$")
plt.plot(xaxisfigure21, yaxis(xaxisfigure21,2), color='orange', label = "$m=2$")
plt.plot(xaxisfigure21, yaxis(xaxisfigure21,3), color='green', label = " $m=3$")    
plt.plot(xaxisfigure21, yaxis(xaxisfigure21,4) , color='red', label = " $m=4$")
plt.plot(xaxisfigure21, yaxis(xaxisfigure21,5) , color='purple', label = "$m=5$")

plt.xlabel("Angle")
plt.ylabel("Mean Absolute Error")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()

plt.show()
```
