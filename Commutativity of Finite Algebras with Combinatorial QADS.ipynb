```python
import math
import numpy as np
import matplotlib.pyplot as plt
def try1(x,s,m):
    z = ((math.cos(x*s))**(2*m))*((math.cos(x*s*m)**2))
    return(z)
def try2(x,S,m):
    First = []
    Second = []
    for i in range(S):
        First.append(try1(x,i,m))
        Sum = sum(First)
        result = 1-Sum /((i+1))
        Second.append(result)
    return(Second)
def plotcombinatorial(x,S,m):
    xaxisfigure1 = np.arange(0,51,1)
    yaxisfigure1 = try2(x,S,m)
    return(xaxisfigure1, yaxisfigure1)    
```

```python
plt.plot(xaxisfigure1, try2(0.177710,51,1), color='orange', label = "C-Grover")
plt.plot(xaxisfigure1, try2(0.177710,51,2), color='purple', label = "C-Grover $ \otimes $ C-Grover")
plt.plot(xaxisfigure1, try2(0.177710,51,3), color='red', label = "Combinatorial $m=3$")     
plt.plot(xaxisfigure1, try2(0.177710,51,4) , color='green', label = "Combinatorial $m=4$") 
plt.plot(xaxisfigure1, try2(0.177710,51,10) , color='blue', label = "Combinatorial $m=10$")
plt.title("Combinatorial Gover QADS")
plt.xlabel("Iterations")
plt.ylabel("Detection Probability")
plt.legend()
plt.plot()
plt.show()
```
